```python
import pandas as pd
import numpy as np
```

```python
model1 = pd.read_csv('not_procesing_ExtraTree.csv',)
model1 = model1.drop('index', axis=1)
model2 = pd.read_csv('not_procesing_randomforest.csv')
model2 = model2.drop('index', axis=1)
model3 = pd.read_csv('not_processing_lgbm.csv')
model3 = model3.drop('index', axis=1)
model4 = pd.read_csv('scale_xgboost.csv')
model4 = model4.drop('index', axis=1)
```

# 하드보팅

```python
combined_predictions = pd.concat([model1, model2, model3, model4], axis=1)

# 다수결로 하드 보팅을 수행합니다.
hard_voting = combined_predictions.mode(axis=1).iloc[:, 0]


# 하드 보팅 결과를 출력합니다.
print("하드 보팅 결과:")
print(hard_voting)
```

```python
submission = pd.read_csv('sample_submission.csv')
submission['quality'] = hard_voting
submission.to_csv('hard_vot.csv',index=False)
```

# 소프트보팅

```python
%cd /content/drive/MyDrive/Colab Notebooks/머신러닝1/조별활동/voting_csv
```

```python
combined_predictions = pd.concat([model1, model2, model3, model4], axis=1)
combined_predictions



# 예측값들의 평균 계산
soft_voting = combined_predictions.mean(axis=1)
rounded_soft_voting = soft_voting.round()


# 가장 근접한 값 찾기
for i in range(len(rounded_soft_voting)):
    diff = np.abs(soft_voting[i] - rounded_soft_voting[i])
    closest_value = combined_predictions.iloc[i].values[np.argmin(diff)]
    rounded_soft_voting[i] = closest_value

# 결과 데이터프레임 생성
soft_voting_result = pd.DataFrame({'quality': rounded_soft_voting})

# 소프트 보팅 결과 출력
print("소프트 보팅 결과:")
print(soft_voting_result)

# csv로 만들기
```

```python
%cd /content/drive/MyDrive/Colab Notebooks/머신러닝1/조별활동/와인품질분류/data
```

```python
submission = pd.read_csv('sample_submission.csv')
submission['quality'] = soft_voting_result
submission.to_csv('soft_vot.csv',index=False)
```
